```python
# Import Libraries 

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) 

# Read in the data 

df = pd.read_csv(r'C:\Users\darke\Downloads\movies.csv')
```

```python
# Taking a look at the data

df.head()
```

```python
# Checking and dropping missing data 

df.isnull().sum()

df = df.dropna(how='any', axis = 0)
```

```python
df.dtypes
```

```python
# Removing decimal point from features that don't utilize them by changing data type

df['budget'] = df['budget'].astype('Int64')
df['gross'] = df['gross'].astype('Int64')
df['votes'] = df['votes'].astype('Int64')
df['runtime'] = df['runtime'].astype('Int64')
```

```python
# Year and Release do not always commute, creating the correct year column

df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]

df.head()
```

```python
df = df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Checking for duplicates, possibly drop duplicates

df['name'].drop_duplicates().sort_values(ascending= False).head()

# I attached head() at the end to suppress output for github, it can be removed at will by anyone
```

```python
df.head()
```

```python
# Scatter plot with Gross Earnings vs Budget

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Gross Earnings vs Budget')

plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Plot Gross Earnings vs Budget using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws= {"color": "red"}, line_kws = {"color": "blue"})
```

```python
# Creating a correlation matrix including only numeric features

df.corr(method = 'pearson')
```

```python
correlation_matrix = df.corr(method = 'pearson') 

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Looking at correlation for company, which is a string 

df.head()
```

```python
df_numerized = df 

for col_name in df_numerized.columns: 
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
        
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson') 

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.50]

high_corr 
```
